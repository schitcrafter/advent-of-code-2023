```python
f = open("3-input.txt", "r")
input = f.read()
print(input)
```

```python
lines = input.split("\n")

specialCharsPos = []
for rowIndex, line in enumerate(lines):
    for colIndex, char in enumerate(line):
        if not char.isdecimal() and not char == ".":
            specialCharsPos.append((rowIndex,colIndex))

print(specialCharsPos)
```

```python
total = 0

currentNumber = 0
startCol = 0
for rowIndex, line in enumerate(lines):
    line += "."
    print(f"Line {rowIndex}: '{line}'")
    for colIndex, char in enumerate(line):
        if char.isdecimal():
            currentNumber *= 10
            currentNumber += int(char)
            # set startCol
            if startCol == 0:
                startCol = colIndex
        else:
            if currentNumber == 0:
                continue

            endCol = colIndex - 1

            print(f"Found integer {currentNumber} going from {rowIndex}:{startCol} to {rowIndex}:{endCol}")

            # currentNumber only counts if there's a special character
            # in a region around the integer
            foundSpecialChar = False
            for specialCharRow, specialCharCol in specialCharsPos:
                if ((rowIndex - 1) <= specialCharRow) and (specialCharRow <= (rowIndex + 1)) \
                    and ((startCol - 1) <= specialCharCol) and (specialCharCol <= (endCol + 1)):
                    foundSpecialChar = True
                    print(f"Found special char for {currentNumber=}, at {specialCharRow}:{specialCharCol}")
                    print(f"Row: {rowIndex-1} <= {specialCharRow} <= {rowIndex+1}")
                    print(f"Col: {startCol-1} <= {specialCharCol} <= {endCol+1}")
                    # break
            
            if foundSpecialChar:
                total += currentNumber
            else:
                print("Didn't find special char")

            # end of integer
            currentNumber = 0
            startCol = 0

print(f"Total: {total}")
```
