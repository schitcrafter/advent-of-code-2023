```rust
:opt 3

use std::fs;

let input: String = fs::read_to_string("8-input.txt").unwrap();
let lines: Vec<String> = input.lines().map(|s| s.to_owned()).collect();

lines
```

```rust
let instructions: Vec<_> = lines[0].chars().map(|instr| instr == 'L').collect();

let mut starts = vec![];
let mut endings = vec![];
let mut ends_with_z = vec![];

let mut current_nodes = vec![];

for (index, line) in lines[2..].iter().enumerate() {
    let line_parts: Vec<_> = line.split("=").collect();
    let start = line_parts[0].trim().to_owned();
    let ending = line_parts[1].trim();

    let ending: Vec<_> = ending[1..ending.len() - 1]
        .split(",")
        .map(|s| s.trim())
        .collect();
    
    endings.push((ending[0].to_string(), ending[1].to_string()));
    ends_with_z.push(start.ends_with("Z"));
    if start.ends_with("A") {
        current_nodes.push(index);
    }
    starts.push(start);
};

let mut ending_indices = vec![];

for (left, right) in &endings {
    let left_index = starts.iter().position(|s| s == left).unwrap();
    let right_index = starts.iter().position(|s| s == right).unwrap();
    ending_indices.push((left_index, right_index));
}

current_nodes
```

```rust
let mut steps: u64 = 0;

for turn_left in instructions.iter().cycle() {
    if current_nodes.iter().all(|node| ends_with_z[*node] == true) {
        break;
    }

    current_nodes = current_nodes.iter().enumerate().map(|(index, current_node)| {
        let possible_paths = ending_indices[*current_node];
        // println!("Got paths {possible_paths:?} for node {current_node}");
    
        let new_node = if *turn_left { possible_paths.0 } else { possible_paths.1 };
        new_node
    }).collect();

    steps += 1;
    if steps % 100_000_000 == 0 {
        println!("million steps: {}", steps / 1_000_000);
    }
}

println!("Total steps taken: {steps}");
```
