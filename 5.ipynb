```python
input_file_name = "5-input.txt"
f = open(input_file_name, "r")
input = f.read()
lines = input.split("\n")
```

```python
# parses a list of numbers (as strings)
# separated by whitespaces to a list of actual numbers
def parse_numbers(string):
    numbers = string.strip().split()
    numbers = list(map(int, numbers))
    return numbers

def parse_mapping(string):
    numbers = parse_numbers(string)
    [dest_range_start, source_range_start, range_len] = numbers
    return (dest_range_start, source_range_start, range_len)

def map_number_single_mapping(number, mapping):
    [dest_range_start, source_range_start, range_len] = mapping

    if source_range_start <= number <= (source_range_start + range_len - 1):
        # number is in source range
        offset = number - source_range_start
        return dest_range_start + offset
    else:
        return None # could not be mapped
    
def map_number_on_map(number, mappings):
    for mapping in mappings:
        possible_mapped_number = map_number_single_mapping(number, mapping)
        if possible_mapped_number != None:
            return possible_mapped_number
        
    return number

map_number_on_map(53, [(50, 98, 2), (52, 50, 48)])
```

```python
line_index = 3 # start out at numbers of first map (4rd line)

maps = []

while line_index < len(lines):
    # collect map
    map_strings = []
    while lines[line_index] != "":
        map_strings.append(lines[line_index])
        line_index += 1
    
    # found empty line at line_index, map ended
    
    line_index += 2 # skip over newline and name of next map

    print(f"{map_strings=}")
    map_numbers = list(map(parse_mapping, map_strings))
    print(f"{map_numbers=}")
    maps.append(map_numbers)

    # seeds = list(map(lambda seed: map_number_on_map(seed, map_numbers), seeds))

    # print(f"Mapped seeds: {seeds}")
    print()
    
```

```python
from multiprocessing import Pool

seeds_unparsed = lines[0].split(":")[1]
seeds = parse_numbers(seeds_unparsed)

def process_seed(seed):
    temp_seed = seed
    for mapping in maps:
        temp_seed = map_number_on_map(temp_seed, mapping)
    return temp_seed

current_min = 10**100

for index in range(0, len(seeds), 2):
    print(f"Seed index: {index} / {len(seeds)}")
    [seed_range_start, seed_range_len] = seeds[index:index + 2]
    seed_range = range(seed_range_start, seed_range_start + seed_range_len)

    with Pool(4) as p:
        print(f"Mapping a total of {seed_range_len} seeds")
        temp_seeds = p.map(process_seed, seed_range)
        print("Calculating minimum")
        min_temp_seeds = min(temp_seeds)
        if min_temp_seeds < current_min:
            current_min = min_temp_seeds
            print(f"Found new minimum: {current_min}")

    # for seed in seed_range:
    #     temp_seed = process_seed(seed)
        
    #     if temp_seed < current_min:
    #         current_min = temp_seed

print(f"Found min: {current_min}")

with open("output.txt", "a") as text_file:
    text_file.write(f"Output for {input_file_name}: {current_min}\n")

# took 102min
```
